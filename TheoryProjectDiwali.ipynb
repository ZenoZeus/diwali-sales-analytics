* EDA

```python
diwaliDf.columns
```

```python
import seaborn as sns
```

```python
# gender
sns.countplot(data=diwaliDf,x="Gender")
```

```python
ax=sns.countplot(data=diwaliDf,x="Gender")
for i in ax.containers:
    ax.bar_label(i)
```

```python
diwaliDf.value_counts().groupby("Gender").sum().plot(kind="bar")
d=diwaliDf.value_counts().groupby("Gender").sum()
d

```

```python
# amount spend by male and female
amt=diwaliDf.groupby("Gender").sum().Amount
```

```python
amt
```

```python
amt.plot(kind='bar')
```

```python
# age
sns.countplot(data=diwaliDf,x='Age Group')
```

```python
ax=sns.countplot(data=diwaliDf,x='Age Group' , hue="Gender")
for i in ax.containers:
    ax.bar_label(i)
```

* order

```python
f=diwaliDf.groupby("State").Orders.sum().sort_values(ascending=True)
f
```

```python
f.plot(kind="bar")
```

* money power vs order

```python
f=diwaliDf.groupby("State").Amount.sum().sort_values()
f.plot(kind="bar")
```

```python
a=diwaliDf.groupby(["Marital_Status","Gender"],as_index=False).Amount.sum()
a
```

```python
sns.barplot(data=a,x="Marital_Status",y="Amount",hue="Gender")
```

```python
sns.countplot(data=diwaliDf,x="Occupation")
plt.rcParams["figure.figsize"]=(16,6)
```

* occupation

```python
ax=sns.countplot(data=diwaliDf,x="Occupation")
for i in ax.containers:
    ax.bar_label(i)
```

```python
s=diwaliDf.groupby("Occupation").Amount.sum().sort_values(ascending=False)
s.plot(kind="bar")
```
